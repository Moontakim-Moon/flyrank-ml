# Quick look at the data

The starter dataset contains 30,000 pages and 44 columns.

The correlation between search_volume and impressions_90d is about 0.001, showing almost no relationship in this dataset.

The median word count is similar across trend categories (around 2,700–2,900 words), suggesting that content length alone may not explain why pages gain or lose traffic.

These observations suggest that multiple signals are needed to identify pages that should be reviewed.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Rows:", len(df))
print("Columns:", len(df.columns))

corr = df["search_volume"].corr(df["impressions_90d"])
print("Correlation:", round(corr, 3))

print("\nMedian word count by trend_direction:")
print(df.groupby("trend_direction")["word_count"].median())
```
